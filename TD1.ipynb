```python
path = "DataProjets.xlsx"
```

## item a

```python
tickersMapping = pd.read_excel(path, sheet_name="Mapping")

def get_msft(df, sedol):
    """Prend en argument un sedol et renvoie le msft correspondant"""
    n = len(df)
    for i in range(n):
        if str(df["Sedol"][i]) == str(sedol):
            return yf.Ticker(df["Tickers"][i])
```

```python
def get_history(msft):
    #return msft.history(period='20y')['Close']
    history = msft.history(start="2003-01-01", end="2022-02-15", interval="1d")['Close']
    #history.rename(index={history.index[0] : history.index[0] + timedelta(days=1)},inplace=True)
    return history

history = pd.DataFrame()

for c in tickersMapping['Sedol']:
    msft = get_msft(tickersMapping, c)
    if not msft == None:
        history[c] = get_history(msft)

        
history
```

```python
def get_mktCap(msft):
    history = msft.history(start="2003-01-01", end="2022-02-15", interval="1d")
    close = history['Close']
    volume = history['Volume']
    return close*volume

mktCaps = pd.DataFrame()

for c in tickersMapping['Sedol']:
    msft = get_msft(tickersMapping, c)
    if not msft == None:
        mktCaps[c] = get_mktCap(msft)

mktCaps
```

```python
marketShare = pd.read_excel(path, sheet_name="MarketCaps", index_col=0)
marketShare
```

## item c

```python
idx = pd.date_range(history.index[0], history.index[-1])
history = history.reindex(idx, fill_value=0)
```

```python
def nettoyage(df2):
    n = len(df2)
    columns = df2.columns
    for column in columns:
        for i in range(n):
            if np.isnan(df2[column][i]) or df2[column][i] == 0:
                if np.sum(df2[column][:i])>0 and np.sum(df2[column][i:])>0:
                    df2[column][i] = df2[column][i - 1]

    return df2
```

```python
historyPrice = nettoyage(history.fillna(0))
historyPrice = historyPrice.iloc[1:]
historyPrice.to_csv('historyPrice.csv')
historyPrice
```

```python
idx = pd.date_range(marketShare.index[0], marketShare.index[-1], freq='M')
marketShare = marketShare.reindex(idx, fill_value=0)
marketShare = nettoyage(marketShare)
marketShare.to_csv('marketShare.csv')
marketShare
```

```python
idx = pd.date_range(mktCaps.index[0], mktCaps.index[-1], freq='M')
marketCaps = mktCaps.reindex(idx, fill_value=0)
marketCaps = nettoyage(marketCaps)
marketCaps.to_csv('marketCaps.csv')
marketCaps
```

## item d

```python
def get_rit(date1, date2, share): 
    """Calcule r_i_t entre une date1 et une date 2 pour un msft"""
    P_i_t = historyPrice[share][date1]
    P_i_tbefore = historyPrice[share][date2]
    if (P_i_tbefore) == 0:
        return 0
    else:
        return (P_i_t / P_i_tbefore) - 1


def get_performance_indice(t=100, N=300):
    """Calcule rt"""
    rt = 0
    date1, date2 = historyPrice.index[t], historyPrice.index[t - 1]
    columns = historyPrice.columns
    for i in range(N):
        companyWeights = marketCaps[columns[i]]
        t_ = datetime(date1.year, date1.month, 1) - timedelta(days=1)
        w_t_i = companyWeights[t_]        
        r_t_i = get_rit(date1, date2, columns[i])
        rt += w_t_i * r_t_i
    return rt

r={}
for date_t in range(len(historyPrice.index)):
    r[date_t] = get_performance_indice(t=date_t, N=300)
#r = pd.DataFrame(r)
#r.index = historyPrice.index
```

```python
r = pd.DataFrame.from_dict(r, orient='index', columns=['r'])
r.index = historyPrice.index
r
```
